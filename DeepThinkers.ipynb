```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from functools import reduce
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv("/content/spain_cities_dataset.csv")
df
```

```python
def first_digit(x):
    try:
        x = abs(float(x))
        cleaned = re.sub(r'[^\d]', '', f"{x:.15g}")
        return int(cleaned[0]) if cleaned else None
    except:
        return None


```

```python
def lat(df):
    df["lat Leading"] = df["lat"].apply(first_digit)
    df = df.dropna(subset=["lat Leading"])
    proportions = df["lat Leading"].value_counts(normalize=True).sort_index()
    return proportions

x=lat(df)
plt.bar(x.index,x.values,color="r")
plt.title("Leading Digit Frequency")
plt.xticks(x.index)
plt.ylabel("Frequency")
plt.xlabel("Leading Digit")
plt.show()
```

```python
def lng(df):
    df["lng Leading"] = df["lng"].apply(first_digit)
    df = df.dropna(subset=["lng Leading"])
    proportions = df["lng Leading"].value_counts(normalize=True).sort_index()
    return proportions
x=lng(df)
plt.bar(x.index,x.values,color="yellow")
plt.title("Leading Digit Frequency")
plt.xticks(x.index)
plt.ylabel("Frequency")
plt.xlabel("Leading Digit")
plt.show()
```

```python
def population(df):
    df["population Leading"] = df["population"].apply(first_digit)
    df = df.dropna(subset=["population Leading"])
    proportions = df["population Leading"].value_counts(normalize=True).sort_index()
    return proportions

x=population(df)
plt.bar(x.index,x.values,color="g")
plt.title("Leading Digit Frequency")
plt.xticks(x.index)
plt.ylabel("Frequency")
plt.xlabel("Leading Digit")
plt.show()
```

```python
def population_proper(df):
    df["population_proper Leading"] = df["population_proper"].apply(first_digit)
    df = df.dropna(subset=["population_proper Leading"])
    proportions = df["population_proper Leading"].value_counts(normalize=True).sort_index()
    return proportions
population_proper(df)
x=population_proper(df)
plt.bar(x.index,x.values,color="purple")
plt.title("Leading Digit Frequency")
plt.xticks(x.index)
plt.ylabel("Frequency")
plt.xlabel("Leading Digit")
plt.show()
```

```python
y = pd.concat([lat(df),lng(df),population(df),population_proper(df)], axis=1).sum(axis=1, skipna=True)
y=y/4
y = y[y.index != 0]
plt.bar(y.index, y.values, color='lavender', label='Bar')
plt.title("Leading Digit Frequency")
plt.xticks(y.index)
plt.ylabel("Frequency")
plt.xlabel("Leading Digit")

plt.show()
```
